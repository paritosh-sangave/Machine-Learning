## Encoding

- OneHotEncoder (sklearn)
- get_dummies (pandas)

```python
import os
os.chdir(r"C:\CDAC\6_Practical_Machine_Learning\Datasets")
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
telecom = pd.read_csv("Telecom.csv")
telecom
```

```python
dum_tel = pd.get_dummies(telecom, drop_first=True)
dum_tel
```

```python
ohe = OneHotEncoder(drop='first')
```

```python
tel_columns = telecom.columns
tel_hot_encoded = pd.DataFrame(ohe.fit_transform(telecom).toarray(), columns = tel_columns)
tel_hot_encoded
```

## Missing Values

```python
from sklearn.impute import SimpleImputer
```

```python
job = pd.read_csv("JobSalary2.csv")
job
```

```python
# Columns with missing values
pd.DataFrame(job.isnull().sum())
# job.isna().sum() ## gives same output
```

```python
pd.DataFrame(job.isnull().sum()>0)
```

```python
# Removing records with NA
job.dropna()
```

```python
#### Constant Imputation
imputer = SimpleImputer(strategy='constant', fill_value = 50)
columns1 = job.columns
pd.DataFrame(imputer.fit_transform(job), columns = columns1)
```

```python
#### Mean Imputation
imputer = SimpleImputer(strategy='mean')
pd.DataFrame(imputer.fit_transform(job), columns = columns1)
```

```python
#### Median Imputation
imputer = SimpleImputer(strategy='median')
pd.DataFrame(imputer.fit_transform(job), columns = columns1)
```

## Scaling Data

1) Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
milk = pd.read_csv("milk.csv", index_col = 0)
milk
```

```python
std_scaler = StandardScaler()
```

```python
std_scaler.fit(milk)
print(std_scaler.mean_)
print(std_scaler.scale_)
```

```python
milk_scaled = std_scaler.transform(milk)
milk_scaled
```

```python
milk_df = pd.DataFrame(milk_scaled, columns = milk.columns, index = milk.index)
milk_df
```

2. Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm_scaler = MinMaxScaler()
```

```python
milk_mm_scaled = mm_scaler.fit_transform(milk)
milk_mm_scaled
```

```python
milk_df2 = pd.DataFrame(milk_mm_scaled, columns = milk.columns, index = milk.index)
milk_df2
```

3. Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust_scaler = RobustScaler()
```

```python
milk_df3 = pd.DataFrame(robust_scaler.fit_transform(milk), columns = milk.columns, index = milk.index)
milk_df2
```
